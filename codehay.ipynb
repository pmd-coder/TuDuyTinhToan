#### Reverse

```python
def doix(a):
    doixung=int((str(a)[::-1]))
    return doixung
def check(a):
    if a!=doix(a):
        return True
    return False
n=int(input())
dem=0
while check(n):
    n=n+doix(n)
    dem=dem+1
print(n, dem)
```

```python
def doixung(n):
    goc=n
    doix=0
    while n>0:
        digit=n%10
        doix=doix*10+digit
        n//=10
    return doix
def kiemtra(n):
    if n != doixung(n):
        return True
    return False
a=int(input())
dem=0
while kiemtra(a):
    a=a+doixung(a)
    dem=dem+1
print(a, dem)
```
