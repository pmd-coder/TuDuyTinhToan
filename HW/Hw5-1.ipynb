Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyen x


```python
def check_amstrong(x):
    temp=x
    tong=0
    n=len(str(x))
    while x>0:
        a=x%10
        x=x//10
        tong=tong+a**n
    if temp==tong:
        return True
    return False
z=int(input())
print(check_amstrong(z))
```
