W4A24 
Nhập vào A, tìm n nhỏ nhất sao cho 
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 

```python
a=float(input())
tong=1
dem=2
while tong<=a:
    tong=tong+1/dem
    dem+=1
print(dem-1)

```

W4A25 
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. 
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
a=int(input())
max_vl=a
min_vl=a
while a!=-1:
    if a>max_vl:
        max_vl=a
    if a<min_vl:
        min_vl=a
    a=int(input())
print(max_vl, min_vl)

```

```python
danh_sach_so = []

while True:
    num = int(input("Nhập số (nhập -1 để dừng): "))
    if num == -1:
        break
    danh_sach_so.append(num)
if len(danh_sach_so) == 0:
    print("Bạn chưa nhập số nào.")
else:
    print(f"\nSố lớn nhất trong dãy: {max(danh_sach_so)}")
    print(f"Số nhỏ nhất trong dãy: {min(danh_sach_so)}")
```
