#### Bài 8: Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này, tính tiền điện tháng này cho hộ biết:
- Từ 0 – 50kWh: 1.984 đồng/kWh
- Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh 


```python
# 1. Nhập dữ liệu đầu vào
ten_chu_ho = input("Nhập tên chủ hộ: ")
chi_so_cu = float(input("Chỉ số tháng trước: "))
chi_so_moi = float(input("Chỉ số tháng này: "))

# 2. Tính số điện tiêu thụ
so_dien = chi_so_moi - chi_so_cu

def tinh_tien_dien(kwh):
    tong_tien = 0
    
    if kwh <= 50:
        tong_tien = kwh * 1984
    elif kwh <= 100:
        tong_tien = 50 * 1984 + (kwh - 50) * 2050
    elif kwh <= 200:
        tong_tien = 50 * 1984 + 50 * 2050 + (kwh - 100) * 2380
    elif kwh <= 300:
        tong_tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (kwh - 200) * 2998
    elif kwh <= 400:
        tong_tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (kwh - 300) * 3350
    else:
        tong_tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (kwh - 400) * 3460
    
    return tong_tien

# 3. Tính toán tiền điện chưa thuế và có thuế
tien_chua_vat = tinh_tien_dien(so_dien)
tien_phai_tra = tien_chua_vat * 1.08  # Thêm VAT 8%

# 4. Xuất kết quả (làm tròn đến đơn vị đồng theo yêu cầu)
print("---")
print(f"Ho va ten: {ten_chu_ho}")
print(f"Tien phai tra la: {int(round(tien_phai_tra, 0))}")
```
