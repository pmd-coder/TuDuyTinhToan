Yêu cầu. Cước phí Taxi Thông Minh.
Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả,
kiểu số nguyên, đơn vị VND).
Các quy tắc cần cài đặt:
1. Giá mở cửa (0–0.8 km):
- Xe 4 chỗ: 11,000 VND
- Xe 7 chỗ: 12,000 VND
2. Giá theo km sau 0.8 km:
- Xe 4 chỗ: từ 0.8→20 km ơnh 13,000 VND/km; trên 20 km tính 10,500 VND/km
- Xe 7 chỗ: từ 0.8→20 km ơnh 14,000 VND/km; trên 20 km tính 12,000 VND/km
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND.
4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường +
chờ).
5. Điều kiện đặc biệt:
- Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền.
- Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu). 

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    a, b, c=0, 0, 0
    nighthour=(22, 23, 0, 1, 2, 3, 4, 5)
    if car_type==4:
        if km<=0.8:
            a=11000*km
        elif 0.8<km<=20:
            a=11000*0.8+(km-0.8)*13000
        else:
            a=11000*0.8+(20-0.8)*13000+(km-20*10500)
    elif car_type==7:
        if km<=0.8:
            a=12000*km
        elif 0.8<km<=20:
            a=12000*0.8+(km-0.8)*14000
        else:
            a=12000*0.8+(20-0.8)*14000+(km-20*12000)
    if wait_minutes<5:
        b=0
    else: 
        b=int(wait_minutes+0.5)*2000
    if start_hour in nighthour:
        c=((a+b)*20)/100
    else:
        c=0
    total=a+b+c
    if km>30 and start_hour not in nighthour:
        total=(total*90)/100
    if total<50000 and km>5:
        total=50000
    return total
a, b, c, d=map(float, input().split())
print(f"số tiền cần trả là {int(smart_taxi_fare(a, b, c, d))}")
    

    
    
```

Yêu cầu: Cho số nguyên x, trả về True nếu tồn tại số nguyên n ≥ 1 sao cho x = 1 + 2 + ... + n, ngược lại trả về False.
Khuyến khích dùng vòng lặp đơn giản để cộng dồn tổng.

```python
x=int(input())
total=0
result="False"
for i in range(1, x):
    total+=i
    if total==x:
        result="True"
        break
print(result)
        
```

Yêu cầu: Cho tọa độ ba điểm A(x1,y1), B(x2,y2), C(x3,y3), trả về True nếu chúng tạo thành một tam giác, ngược lại trả về
False.
Quy tắc: Chúng tạo thành tam giác nếu không thẳng hàng (diện tích có hướng khác 0)

```python
x1, y1=map(float, input().split())
x2, y2=map(float, input().split())
x3, y3=map(float, input().split())
if (x2-x1)*(y3-y1)-(y2-y1)*(x3-x1)!=0:
    print("True")
else:
    print("False")

```
