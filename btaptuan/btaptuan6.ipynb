giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]


```python
a=list(map(int, input().split()))
print(list(dict.fromkeys(a)))

```

```python
a=list(map(int, input().split()))
l=[]
for i in a:
    if i not in l:
        l.append(i)
print(l) 
```

#### W6A2
Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu 
tới vị trí i), giữ nguyên thứ tự


```python
a=list(map(int,input().split()))
res=[]
tong=0
for i in a:
    tong+=i
    res.append(tong)
print(res)

```

```python
a=list(map(int,input().split()))
copy=a[:]
for i in range(len(a)):
    a[i]=sum(copy[:i+1])
print(copy)
```

sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp
xếp.)


```python
a=tuple(map(int,input().split()))
k=int(input())
a=a[k:]+a[:k]
print(a)
```
